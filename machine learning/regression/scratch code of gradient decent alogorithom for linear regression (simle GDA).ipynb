# Step 1 :- Omega(w) and Bais (b)

```python
def best_w_single(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    num = np.sum((x - x_mean) * (y - y_mean))
    den = np.sum((x - x_mean) ** 2)

    w = num / den
    b = y_mean - w * x_mean
    return w, b
```

# Step 2 :- Model creation

```python
def model(w, b, x):
    f_wb = w * x + b
    return f_wb
```

# Step 3 :- compute cost 

```python
def compute_cost(w,x,b,y):
    M = len(x) 
    cost = 0
    for i in range(M):
        f_wb = w*x[i]+b
        cost = cost + (y[i]-f_wb)**2
        total_cost = cost/(2*M)
    return total_cost
        
        
```

```python
import numpy as np

def compute_cost(w, X, b, y):
    f_wb = np.dot(X, w) + b
    return np.sum((f_wb - y)**2) / (2 * X.shape[0])
```

# Step 4 :-  Gradient

```python
def gradient(x,w,b,y):
    m = len(x)
    dj_dw = 0
    dj_db = 0
    for i in range(m):
        f_wb = w*x[i]+b
        dj_dw_i = (y[i]-f_wb) * x[i] / m
        dj_db_i = (y[i]-f_wb)/m

        dj_dw += dj_dw_i
        dj_db += dj_db_i

    return dj_dw, dj_db  
```

# For multivariable 

```python
import numpy as np 
def gradient(x,w,b,y):
    m = x.shape[0]
    dj_dw = 0
    dj_db = 0
    for i in range(m):
        f_wb = np.sum(w*x[i]) + b
        dj_dw += (f_wb - y[i]) * x[i]
        dj_db += (f_wb - y[i])      

    dj_dw /= m                          
    dj_db /= m

    return dj_dw, dj_db
```

# Step 5 :- GDA(Gradient decent algorithom)

```python
def gradient_descent_single(X, Y, alpha=0.01):
    m = len(X)
    w = 0  # initial weight
    b = 0  # initial bias
    
    for i in range(m):
        f_wb = w*X[i] + b
        dw += (f_wb - Y[i]) * X[i]
        db += (f_wb - Y[i])

        dw /= m
        db /= m

        # update weights
        w = w - alpha * dw
        b = b - alpha * db

    return w, b
```
