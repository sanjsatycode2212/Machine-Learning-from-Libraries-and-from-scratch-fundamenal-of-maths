# Step 1 :- Omega(w) and Bais (b)

```python
def best_w_single(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    num = np.sum((x - x_mean) * (y - y_mean))
    den = np.sum((x - x_mean) ** 2)

    w = num / den
    b = y_mean - w * x_mean
    return w, b
```

# Step 2 :- Model creation

```python
def model(w, b, x):
    f_wb = w * x + b
    return f_wb
```

# Step 3 :- compute cost 

# Lasso

```python
def compute_cost_lasso(w, X, b, Y, lambda_):
    M = len(X)
    cost = 0
    for i in range(M):
        f_wb = w*X[i] + b
        cost += (Y[i] - f_wb)**2

    cost = cost/(2*M) + lambda_ * abs(w)
    return cost
```

```python
import numpy as np

def compute_cost_lasso(w, X, b, y, lambda_):
    m = X.shape[0]
    f_wb = np.dot(X, w) + b
    cost = (1 / (2*m)) * np.sum((f_wb - y)**2) + lambda_ * np.sum(np.abs(w))
    return cost
```

```python
def compute_cost_ridge(w, X, b, Y, lambda_):
    M = len(X)
    cost = 0
    for i in range(M):
        f_wb = w*X[i] + b
        cost += (Y[i] - f_wb)**2

    cost = cost/(2*M) + (lambda_/2) * (w)**2
    return cost
```

```python
def gradient(x,w,b,y):
    m = len(x)
    dj_dw = 0
    dj_db = 0
    for i in range(m):
        f_wb = w*x[i]+b
        dj_dw_i = (y[i]-f_wb) * x[i] / m
        dj_db_i = (y[i]-f_wb)/m

        dj_dw += dj_dw_i
        dj_db += dj_db_i

    return dj_dw, dj_db  
```

```python
import numpy as np 
def gradient(x,w,b,y):
    m = x.shape[0]
    dj_dw = 0
    dj_db = 0
    for i in range(m):
        f_wb = np.sum(w*x[i]) + b
        dj_dw += (f_wb - y[i]) * x[i]
        dj_db += (f_wb - y[i])      

    dj_dw /= m                          
    dj_db /= m

    return dj_dw, dj_db
```

```python
# Step 5 :- GDA(Gradient decent algorithom)
```

```python
def gradient_descent_single(X, Y, alpha=0.01):
    m = len(X)
    w = 0  # initial weight
    b = 0  # initial bias
    
    for i in range(m):
        f_wb = w*X[i] + b
        dw += (f_wb - Y[i]) * X[i]
        db += (f_wb - Y[i])

        dw /= m
        db /= m

        # update weights
        w = w - alpha * dw
        b = b - alpha * db

    return w, b
```
