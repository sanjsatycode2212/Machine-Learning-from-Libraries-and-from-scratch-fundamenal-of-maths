```python
import numpy as np
```

```python
x = np.array([0,1,2,3,4,5])

y = np.array([0,0,0,1,1,1])
```

```python
# now we define bais and wheight
w = 0.0
b = 0.0
```

```python
# then using sigmoid function get logistic function
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def predict(x,w,b):
    z = w*x+b
    return sigmoid(z)
```

```python
# now be use binary cross entrophy 
def loss(y,y_pred):
    n = len(y)
    return -np.mean(y*np.log(y_pred)+(1-y)*np.log(1-y_pred))


# n = number of y value
# mean = sigma 
# y = actual output 0 and 1
# y^ = predict output by model
```

```python
def train(x, y, lr=0.1, epochs=1000):
    w, b = 0, 0  # initialize
    n = len(x)
    for i in range(epochs):
        y_pred = predict(x, w, b)
        
        # gradients
        dw = np.dot((y_pred - y), x) / n
        db = np.sum(y_pred - y) / n
        
        # update weights
        w -= lr * dw
        b -= lr * db
        
        if i % 200 == 0:
            print(f"Epoch {i}, Loss: {loss(y, y_pred):.4f}")
    return w, b
```

```python
w, b = train(x, y, lr=0.1, epochs=1000)
print("Learned weight:", w)
print("Learned bias:", b)
```

```python
y_pred_prob = predict(x, w, b)
y_pred_class = (y_pred_prob >= 0.5).astype(int)

print("Predicted Probabilities:", y_pred_prob)
print("Predicted Classes:", y_pred_class)

accuracy = np.mean(y_pred_class == y)
print("Accuracy:", accuracy)
```
