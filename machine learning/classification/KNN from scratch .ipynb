# ** Sample dataset **

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = [[1,2],[2,3],[3,4],[6,7],[7,8],[8,9]] # feature point 
y = [0,0,0,1,1,1] # -----> lables(output)
```

# ** Distance function ** 

```python
def distance(x1,x2):
    return ((x1[0]-x2[0])**2 + (x1[1]-x2[1])**2)**0.5
```

# ** Define KNN Functions **

```python
def Knn_pre(X,y,test_point,k=3):
    dists = []
    # sab points se distance nikalne ke liye
    for i in range(len(X)):
        d = distance(X[i],test_point)
        dists.append((d,y[i]))
    # nearest points uper aa jayenge
    dists.sort()

    labels = []
    for i in range(k):
        labels.append(dists[i][1])

    if labels.count(1)>labels.count(0):
        return 1
    else:
        return 0
```

```python
test = [10,6]
pred = Knn_pre(X,y,test,3)
```

```python
for i in range(len(X)):
    if y[i]==0:
        plt.scatter(X[i][0],X[i][1],marker='o')
    else:
        plt.scatter(X[i][0],X[i][1],marker='s')

# plot test point
plt.scatter(test[0],test[1],marker='x',s=200)

plt.title("KNN Visualization")
plt.show()

print("Prediction =",pred)
```
